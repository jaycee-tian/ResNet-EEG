```python
# 画 100x100 噪声图片
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import matplotlib.image as mpimg
import matplotlib.colors as colors
import matplotlib.patches as patches
import matplotlib.gridspec as gridspec
import matplotlib.animation as animation
from IPython.display import HTML
%matplotlib inline

# 生成 100x100 的随机数
noise = np.random.randn(100, 100)
# 画图
plt.figure(figsize=(5,5))
plt.imshow(noise, cmap='gray')
plt.show()
```

```python
# print %.2f, np.min(noise),np.median(noise),np.max(noise)
print('%.2f %.2f %.2f %.2f' % (np.min(noise),np.median(noise),np.max(noise),np.std(noise)))

print(noise.shape)
print(np.mean(noise))
norm_noise = (noise - np.mean(noise))/np.std(noise)
print('%.2f %.2f %.2f' % (np.min(norm_noise),np.median(norm_noise),np.max(norm_noise)))
# flatten noise
flatten_noise = noise.flatten()
print(flatten_noise)
# 画折线图
plt.figure(figsize=(5,5))
plt.plot(flatten_noise)
plt.show()
```
