```python
import torch

# 假设 x 是一个形状为 [64, 3, 100, 100] 的张量
x = torch.rand(64, 3, 100, 100)

# 将每个样本数据展平为一维
flat_x = x.view(64, -1)

# 对每个样本计算中位数
# medians = torch.median(flat_x, dim=1).values
medians,_ = torch.median(flat_x, dim=1)

print(medians.shape)
```

```python
import torch

# 假设 x 是一个形状为 [64, 3, 100, 100] 的张量
x = torch.rand(64, 3, 100, 100)

# 定义一个条件函数，例如计算每个样本的平均值
sample_mean = x.mean(dim=(1, 2, 3))

# 定义筛选条件，例如平均值大于0.5的样本
mask = sample_mean > 0.5

# 使用布尔掩码筛选满足条件的样本
filtered_samples = x[mask]

print(filtered_samples.shape)
```
